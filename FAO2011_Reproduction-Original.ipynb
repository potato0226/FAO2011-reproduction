```python
dt5=dt4[(dt4["AggItem"].isin([1,2,3,4,5,6,7]))&(dt4["AggArea"].isin([1,2,3,4,5,6,7]))]

#edible production
dt5["ProductionE"] = dt5["Production"] * dt5["AllocationF"] * dt5["ConversionF"]

#Production Loss 
dt5["ProductionLoss"]=(dt5["ProductionLossR"]/(100-dt5["ProductionLossR"]))*dt5["Production"]
#dt5["ProductionLoss"]=(dt5["ProductionLossR"]/(100-dt5["ProductionLossR"]))*dt5["Production"] * dt5["AllocationF"]

#Postharvest Loss
dt5["PostharvestLoss"]=dt5["PostharvestLossR"]/100*dt5["Production"]
#dt5["PostharvestLoss"]=dt5["PostharvestLossR"]/100*dt5["Production"] * dt5["AllocationF"]

#Processing Loss
dt5["ProcessingLoss"]=dt5["ProcessingLossR"]/100*(dt5["Processing"]+(1-dt5["FreshF"])*dt5["Food"])
#Distribution Loss (Simple)
dt5["DistributionLoss"]=dt5["DistributionLossR"]/100*(dt5["Food"]-dt5["ProcessingLoss"])
#Consumption Loss (Simple)
dt5["ConsumptionLoss"]=dt5["ConsumptionLossR"]/100*(dt5["Food"]-dt5["DistributionLoss"]-dt5["ProcessingLoss"])

## edible loss
#dt5["ProductionLossE"]=dt5["ProductionLoss"]*dt5["ConversionF"]
#dt5["PostharvestLossE"]=dt5["PostharvestLoss"]*dt5["ConversionF"]
dt5["ProductionLossE"]=dt5["ProductionLoss"]*dt5["ConversionF"]*dt5["AllocationF"]
dt5["PostharvestLossE"]=dt5["PostharvestLoss"]*dt5["ConversionF"]*dt5["AllocationF"]

dt5["ProcessingLossE"]=dt5["ProcessingLoss"]*dt5["ConversionF"]
dt5["DistributionLossE"]=dt5["DistributionLoss"]*dt5["ConversionF"]
dt5["ConsumptionLossE"]=dt5["ConsumptionLoss"]*dt5["ConversionF"]

#total loss
dt5["TotalLoss"]=dt5["ProductionLoss"]+dt5["PostharvestLoss"]+dt5["ProcessingLoss"]+dt5["DistributionLoss"]+dt5["ConsumptionLoss"]
dt5["TotalLossE"]=dt5["ProductionLossE"]+dt5["PostharvestLossE"]+dt5["ProcessingLossE"]+dt5["DistributionLossE"]+dt5["ConsumptionLossE"]


dt5.head()
```
